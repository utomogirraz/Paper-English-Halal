```python
query="ketan bawang"
qlist=query.lower().split()
query="|".join(qlist)
query
```

```python
mask = datahalal[datahalal['nama'].str.contains(query) | datahalal['produsen'].str.contains(query,case=False)]
mask
```

```python
mask=mask.reset_index()
```

```python
def jaccard_similarity(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return float(len(s1.intersection(s2)) / len(s1.union(s2)))
```

```python
def get_response(q):
    similarity=np.zeros((len(mask)))
    for ind in mask.index:
        nama_produk = str(mask['nama'][ind])
        nama_produsen = str(mask['produsen'][ind])
        line=nama_produk+' '+nama_produsen
        if (type(line)==str):
            qlist=q.lower().split()
            linelist=line.lower().split()
            similarity[ind]=jaccard_similarity(qlist,linelist)
    ##idx = np.argmax(similarity)
    ##if(similarity[idx]<=0.4):
        ##idx=2083
    idx=(-similarity).argsort()[:1]
    return(idx)
```

```python
def get_response1(q):
    similarity=np.zeros((len(mask)))
    for ind in mask.index:
        nama_produk = str(mask['nama'][ind])
        nama_produsen = str(mask['produsen'][ind])
        line=nama_produk+' '+nama_produsen
        if (type(line)==str):
            qlist=q.lower().split()
            linelist=line.lower().split()
            similarity[ind]=jaccard_similarity(qlist,linelist)
    ##idx = np.argmax(similarity)
    idx=(-similarity).argsort()[:1]
    return(similarity[idx])
```

```python
def get_response2(q):
    similarity=np.zeros((len(mask)))
    for ind in mask.index:
        nama_produk = str(mask['nama'][ind])
        nama_produsen = str(mask['produsen'][ind])
        line=nama_produk+' '+nama_produsen
        if (type(line)==str):
            qlist=q.lower().split()
            linelist=line.lower().split()
            similarity[ind]=jaccard_similarity(qlist,linelist)
    idx = np.argmax(similarity)
    ##if(similarity[idx]<=0.4):
        ##idx=2083
    ##idx=(-similarity).argsort()[:1]
    return(idx)
```

```python
q="Aqua Air Mineral 1500Ml"
nums=get_response(q)
items=[]
##q3 = get_response1(q)    
##data.append(q3)
for y in nums:
    items.append([mask['nama'][y], mask['produsen'][y], mask['sertif'][y],mask['tanggal'][y]])
##data.append(items)
```
