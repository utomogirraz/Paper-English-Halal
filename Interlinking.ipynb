```python
import numpy as np
import pandas as pd
from rdfpandas.graph import to_dataframe
import rdflib
```

```python
dfjaccard = pd.read_csv('Jaccard_labeling_benar.csv')
dfjaccard.drop(['Column1'],inplace=True, axis=1)
dfjaccard.drop(['TrueFalse'],inplace=True, axis=1)
dfjaccard
```

```python
dfann = pd.read_csv('ANN_Similarity_Neo4j.csv')
dfann
```

```python
dfann['LabelingANN'] = np.where(dfann['similarity'] >= 0.42, 1, 0)
dfann
```

```python
dfaa = pd.read_csv('Adamic_Adar_Link_Prediction_Neo4j.csv')
dfaa
```

```python
dfaa['LabelingAA'] = np.where(dfaa['score'] >= 0.42, 1, 0)
dfaa
```

```python
dfpa = pd.read_csv('Preferential_Attachment_Link_Neo4j.csv')
dfpa
```

```python
dfpa['LabelingAA'] = np.where(dfpa['score'] >= 0.42, 1, 0)
dfpa
```

```python
dfann.to_csv(r'Labeling_ANN_benar.csv')
dfaa.to_csv(r'Labeling_AA_benar.csv')
dfpa.to_csv(r'Labeling_PA_benar.csv')
```

# Jaccard Similarity

```python
dfann = pd.merge( dfann,dfjaccard, left_on=['from','to'], right_on=['from','to'])
dfann
```

```python
dfann['TrueFalse'] = np.where(dfann['LabelingANN'] == 1.0, np.where(dfann['Labeling'] == 1.0, '1', '0'), '0')
dfann
```

```python
dfann.to_csv(r'Labeling_ANN_benar.csv')
```

```python
index_names = dfann[ dfann['TrueFalse'] == '0' ].index
dfann.drop(index_names, inplace = True)
dfann
```

```python
dfaa = pd.merge( dfaa,dfjaccard, left_on=['from','to'], right_on=['from','to'])
dfaa
```

```python
dfaa['TrueFalse'] = np.where(dfaa['LabelingAA'] == 1.0, np.where(dfaa['Labeling'] == 1.0, '1', '0'), '0')
dfaa
```

```python
index_names = dfaa[ dfaa['TrueFalse'] == '0' ].index
dfaa.drop(index_names, inplace = True)
dfaa
```
